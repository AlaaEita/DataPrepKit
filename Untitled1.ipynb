```python
import os
from PIL import Image
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Function to count class occurrences in a directory
def count_class_occurrences(directory):
    class_counts = {}
    for root, dirs, files in os.walk(directory):
        for file in files:
            class_label = os.path.basename(root)
            class_counts[class_label] = class_counts.get(class_label, 0) + 1
    return class_counts
```

```python
# Display class distribution
def display_class_distribution(dir, set_name):
    class_counts = count_class_occurrences(dir)
    print(f"{set_name} set class distribution:")
    print(class_counts)

train_dir = "C:/User/Pharmacy/Downloads/train"
test_dir = "C:/Users/Pharmacy/Downloads/test"
val_dir = "C:/Users/Pharmacy/Downloads/val"

display_class_distribution(train_dir, "Training")
display_class_distribution(test_dir, "Testing")
display_class_distribution(val_dir, "Validation")
```

```python
# Function to get image dimensions
def get_image_dimensions(directory):
    image_dimensions = {}
    for root, dirs, files in os.walk(directory):
        for file in files:
            image_path = os.path.join(root, file)
            try:
                with Image.open(image_path) as img:
                    width, height = img.size
                image_dimensions[image_path] = (width, height)
            except Exception as e:
                print(f"Error processing image {image_path}: {str(e)}")
    return image_dimensions
```

```python
# Get and display image dimensions
def display_image_dimensions(dir, set_name):
    image_dims = get_image_dimensions(dir)
    print(f"{set_name} set image dimensions:")
    for image_path, dimensions in image_dims.items():
        print(f"{image_path}: {dimensions}")

display_image_dimensions(train_dir, "Training")
display_image_dimensions(test_dir, "Testing")
display_image_dimensions(val_dir, "Validation")
```

```python
from sklearn.model_selection import train_test_split
from skimage import io
import numpy as np
```

```python
# Normalize pixel values
def normalize_images(image_paths):
    normalized_images = {}
    for image_path in image_paths:
        try:
            img = io.imread(image_path)
            normalized_img = img.astype(np.float32) / 255.0
            normalized_images[image_path] = normalized_img
        except Exception as e:
            print(f"Error processing image {image_path}: {str(e)}")
    return normalized_images
```

```python
train_image_dims = get_image_dimensions(train_dir)
test_image_dims = get_image_dimensions(test_dir)
val_image_dims = get_image_dimensions(val_dir)
```

```python
image_paths = list(train_image_dims.keys()) + list(test_image_dims.keys()) + list(val_image_dims.keys())
train_paths, test_paths = train_test_split(image_paths, test_size=0.2, random_state=42)
```

```python
normalized_train_images = normalize_images(train_paths)
normalized_test_images = normalize_images(test_paths)
```

```python
import tensorflow as tf
from tensorflow.keras import layers
from sklearn.metrics import confusion_matrix
import numpy as np
```

```python
# Mock data
train_images = np.random.rand(1000, 224, 224, 3)
train_labels = np.random.randint(2, size=(1000, 1))
val_images = np.random.rand(200, 224, 224, 3)
val_labels = np.random.randint(2, size=(200, 1))
test_images = np.random.rand(200, 224, 224, 3)  # Adjust size if needed
test_labels = np.random.randint(2, size=(200, 1))  # Ensure dimensions match test_images
```

```python
#Normalize data
normalized_train_images = train_images.astype('float32') / 255.0
normalized_val_images = val_images.astype('float32') / 255.0
normalized_test_images = test_images.astype('float32') / 255.0
train_labels = train_labels.astype('float32')
val_labels = val_labels.astype('float32')
test_labels = test_labels.astype('float32')  # Normalize if needed
```

```python
# Define CNN model
def build_cnn_model():
    model = tf.keras.models.Sequential([
        tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(1, activation='sigmoid')
    ])
    return model
```

```python
# Hyperparameter tuning
def train_and_evaluate_model(model, train_images, train_labels, val_images, val_labels, batch_size=32, epochs=10):
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    history = model.fit(train_images, train_labels, batch_size=batch_size, epochs=epochs,
                        validation_data=(val_images, val_labels))
    return history
```

```python
# Model evaluation
def evaluate_model(model, test_images, test_labels):
    loss, accuracy = model.evaluate(test_images, test_labels)
    predictions = model.predict(test_images)
    predicted_labels = np.round(predictions).astype(int)
    cm = confusion_matrix(test_labels, predicted_labels)
    return loss, accuracy, cm
```

```python
# Example experiment
results = []
num_experiments = 5
for i in range(num_experiments):
    print(f"Experiment {i+1}/{num_experiments}")
    model = build_cnn_model()
    history = train_and_evaluate_model(model, normalized_train_images, train_labels, normalized_val_images, val_labels)
    loss, accuracy, cm = evaluate_model(model, normalized_test_images, test_labels)
    results.append({
        'experiment': i + 1,
        'loss': loss,
        'accuracy': accuracy,
        'history': history,
        'confusion_matrix': cm
    })
```

```python
# Analyze and visualize the results
for result in results:
    print(f"Experiment {result['experiment']}:")
    print("Test set loss:", result['loss'])
    print("Test set accuracy:", result['accuracy'])
    cm_df = pd.DataFrame(result['confusion_matrix'], index=['Normal', 'Pneumonia'], columns=['Predicted Normal', 'Predicted Pneumonia'])
    sns.heatmap(cm_df, annot=True, fmt="d", cmap="Blues")
    plt.title(f"Confusion Matrix - Experiment {result['experiment']}")
    plt.show()
```
