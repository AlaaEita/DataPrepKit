#Handling Missing Values:

```python
def check_null_values(df):
    return df.isnull()
```

```python
null_values = check_null_values(df)
print(null_values)
```

```python
df.isnull()
```

```python
def fill_null_values(df, value):
    return df.fillna(value=value)
```

```python
df2 = fill_null_values(df, value=0)
print(df2)
```

```python
df2=df.fillna(value=0)
df2
```

```python
df2.isnull().sum().sum()
```

```python
def fill_null_with_value(df, fill_value):
    return df.fillna(value=fill_value)
```

```python
fill_value = 5  # Replace 5 with the value you want to fill null values with
df3 = fill_null_with_value(df, fill_value)
print(df3)
```

```python
df3=df.fillna(value=5)
df3
```

```python
df3.isnull().sum()
```

#filling Null Values with a previous value

```python
def fill_null_forward(df):
    return df.fillna(method='pad')
```

```python
df4 = fill_null_forward(df)
print(df4)
```

```python
df4=df.fillna(method='pad')
df4
```

```python
df4.isnull().sum()
```

#filling Null Values with a nextvalue

```python
def fill_null_backward(df):
    return df.fillna(method='bfill')
```

```python
df5 = fill_null_backward(df)
print(df5)
```

```python
df5=df.fillna(method='bfill')
df5
```

```python
def fill_null_forward_row(df):
    return df.fillna(method='pad', axis=1)
```

```python
df6 = fill_null_forward_row(df)
print(df6)
```

```python
df6=df.fillna(method='pad',axis=1)
df6
```

```python
def fill_null_backward_row(df):
    return df.fillna(method='bfill', axis=1)
```

```python
df7 = fill_null_backward_row(df)
print(df7)
```

```python
df7=df.fillna(method='bfill',axis=1)
df7
```

```python
def fill_null_custom(df, fill_values):
    return df.fillna(value=fill_values)
```

```python
fill_values = {'Age': 'abcd', 'Cabin': 'defg'}
df8 = fill_null_custom(df, fill_values)
print(df8)
```

```python
df8=df.fillna({'Age':'abcd','Cabin':'defg'})
df8
```

#filling null values with the mean of a columns

```python
def fill_null_with_mean(df, column_name):
    return df.fillna(value=df[column_name].mean())
```

```python
column_name = 'Age'  # Replace 'Age' with the column name you want to fill null values with its mean
df9 = fill_null_with_mean(df, column_name)
print(df9)
```

```python
df9=df.fillna(value=df['Age'].mean())
df9
```

```python
df9=df.fillna(value=df['Age'].max())
df9
```

```python
df9=df.fillna(value=df['Age'].min())
df9
```

```python
def drop_null_rows(df):
    return df.dropna()
```

```python
df10 = drop_null_rows(df)
print(df10)
```

```python
df10=df.dropna()
df10
```

#replace

```python
def replace_nan(df, replacement_value):
    return df.replace(to_replace=np.nan, value=replacement_value)
```

```python
replacement_value = 60  # Replace 60 with the value you want to replace NaN with
df12 = replace_nan(df, replacement_value)
print(df12)
```

```python
df12=df.replace(to_replace=np.nan,value=60)
df12
```
